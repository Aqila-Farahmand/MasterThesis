```python
import re
from string import punctuation
import numpy as np
```

```python
df1["text_str"] = df1['text'].astype(str)
```

```python
df1
```

```python
def preprocess_text(text):
    text = text.lower()  # Lowercase text
    text = re.sub(f"[{re.escape(punctuation)}]", "", text)  # Remove punctuation
    text = " ".join(text.split())  # Remove extra spaces, tabs, and new lines
    text = re.sub(r"https?://\S+", "", text)
    return text

df1["text_clean"] = df1["text_str"].map(preprocess_text)
```

```python
df1
```

```python
df2 = df1[["text_clean" , "label"]]
```

```python
df2
```

```python
df2.to_csv("commit_dataset_clean.csv")
```

```python
df2
```

```python

# Step 2: Filter rows where the length of text in 'text_column' is at least 10 characters
filtered_data = df2[df2['text_clean'].str.len() >= 10]

# Step 3: Save the filtered DataFrame back to a CSV file
filtered_data.to_csv('clean_TD_commits.csv', index=False)
```
